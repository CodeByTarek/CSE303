***Plotting Practice***

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure (figsize=(8,6), dpi = 80)

X = np.linspace(-2*np.pi, 2*np.pi, 500, endpoint = True)
C = np.cos(X)
S = np.sin(X)
T = np.tan(X)

plt.subplot(3,1,1)
plt.plot(X, C, color = "red", label = "cosine", linestyle = "-", linewidth = 2.0)
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])

t = 0
plt.plot([t,t], [-1, np.cos(t)], color = 'red', linestyle = "--", linewidth = 1.0)
plt.annotate(r'$ \cos0 = 1$', xy = (t, np.cos(t)), xycoords = 'data', xytext = (+100, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .2"))
plt.title("Cosine Curve")

plt.subplot(3,1,2)
plt.plot(X, S, color = "blue", label = "sine", linestyle = "-", linewidth = 2.0)
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])
plt.title("Sine Curve")

plt.subplot(3,1,3)
plt.plot(X, T, color = "green", label = "tan", linestyle = "-", linewidth = 2.0)

plt.xlim(-4.0, 4.0)
#plt.ylim(-1.0, 1.0)

#plt.xticks (np.arange(-4,5,1))
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])
#plt.yticks (np.arange(-1,1.5,0.5))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

#plt.legend(loc = "upper left")

plt.title("Tan Curve")
plt.savefig("sample1.png",dpi=72)

plt.subplots_adjust(hspace = 1)
plt.show()
```

**Genearate 4 different datasets with two variables x and y using np.random() function. After that, create a figure with 2*2 grid that shows the scatter plot of these datasets**


```python
d1 = [[np.random.randint(1, 100) for i in range(1, 101)],[np.random.randint(1, 100) for i in range(1, 101)]]
d2 = [[np.random.randint(1, 200) for i in range(1, 201)],[np.random.randint(1, 200) for i in range(1, 201)]]
d3 = [[np.random.randint(1, 300) for i in range(1, 301)],[np.random.randint(1, 300) for i in range(1, 301)]]
d4 = [[np.random.randint(1, 400) for i in range(1, 401)],[np.random.randint(1, 400) for i in range(1, 401)]]

plt.figure (figsize=(8,6), dpi = 80)
plt.subplot(2,2,1)
plt.scatter(d1[0],d1[1], color='red', marker='o')

plt.subplot(2,2,2)
plt.scatter(d2[0],d2[1], color='green', marker='x')

plt.subplot(2,2,3)
plt.scatter(d3[0],d3[1], color='blue', marker='*')

plt.subplot(2,2,4)
plt.scatter(d4[0],d4[1], color='purple', marker='^')
plt.show()
```
